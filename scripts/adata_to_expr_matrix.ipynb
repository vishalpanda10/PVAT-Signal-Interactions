```python
import scanpy as sc
from anndata import AnnData
import pandas as pd
import anndata as ad
```

```python
# loading the original anndata
pvat_cell_atlas_data_final = sc.read_h5ad('../data/PVAT_BAT_SCPAnnotated.h5ad')
pvat_cell_atlas_data_final
```

```python
pvat_cell_atlas_data_final.to_df()
```

```python
# # Normalize the gene expression data to a total count of 10,000 per cell, then apply a log transformation to stabilize variance across cells.
sc.pp.normalize_total(pvat_cell_atlas_data_final, target_sum=1e4)
sc.pp.log1p(pvat_cell_atlas_data_final)
```

```python
pvat_cell_atlas_data_final.to_df()
```

```python
pvat_cell_atlas_data_final.obs['celltype'].unique().tolist()
```

```python
pvat_cell_atlas_data_final.obs['sample_id'].value_counts()
```

```python
pvat_cell_atlas_df = pvat_cell_atlas_data_final.to_df()
pvat_cell_atlas_df.head()
```

```python
#Filtering pvat meta data(not BAT)
pvat_metadata = pvat_cell_atlas_data_final.obs
pvat_metadata = pvat_metadata[(pvat_metadata.loc[:,'organ__ontology_label'] == 'adipose tissue|perivascular space|thoracic aorta')]
pvat_metadata.head()
```

```python
# Filtering PVAT cell by gene matrxix(BAT absent)
pvat_cell_by_gene_matrix = pvat_cell_atlas_df.loc[pvat_metadata.index]
```

```python
pvat_cell_by_gene_matrix.shape
```

```python
pvat_cell_by_gene_matrix.to_csv('../data/pvat_cell_by_gene_matrix.csv')
pvat_metadata.to_csv('../data/pvat_metadata.csv')
```

```python
pvat_cell_by_gene_matrix
```

```python
#Filtering bat meta data
bat_metadata = pvat_cell_atlas_data_final.obs
bat_metadata = bat_metadata[(bat_metadata.loc[:,'organ__ontology_label'] == 'brown adipose tissue')]
bat_metadata.head()
```

```python
# Filtering bat cell by gene matrxix
bat_cell_by_gene_matrix = pvat_cell_atlas_df.loc[bat_metadata.index]
```
